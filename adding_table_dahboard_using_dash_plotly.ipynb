```python
from dash import Dash, dcc, html, Input, Output
import plotly.express as px

# fig = py.get_figure('https://plotly.com/~Dreamshot/8152/', raw=True)
app = Dash(__name__)
params = ['Global_Connected_Services_2021']
# image_filename = r'D:\dashBorad_using_dash\s.jpg' # replace with your own image
# # image_filename="https://raw.githubusercontent.com/cldougl/plot_images/add_r_img/vox.png"
# encoded_image = base64.b64encode(open(image_filename, 'rb').read())
# fig=r'D:\dashBorad_using_dash\s.jpg'
# fig = go.Figure(fig)
app.layout = html.Div([
#      html.Img(src='data:image/jpg;base64,{}'.format(encoded_image),height=50, width=50),
    html.H1(id='Title1',children='Restaurant tips by day of week',style={'text-align': 'center'}),
    dash_table.DataTable(
        id='table-editing-simple',
        columns=(
#            [{'id': 'SN', 'name': 'SN'}] +
            [{'id': p, 'name': p} for p in params]
        ),
        data=[
            dict(**{param: 58  for param in params})
        ],
        style_data={'border': '1px solid blue'},
    style_header={
        'border': '1px solid pink',
        'textAlign': 'center'},
    style_cell={
        'textAlign': 'center'
    },
        editable=True),
    html.Br(),
    dcc.Dropdown(
        id="dropdown",
        options=["R1", "R2", "R3","R4"],
        value="R1",
        clearable=False,
    ),
    dcc.Graph(id="graph"),
#      dcc.Graph(id='table-editing-simple-output'),
])


@app.callback(
    Output("graph", "figure"), 
    Input("dropdown", "value"))
def update_bar_chart(region):
    df = px.data.tips() # replace with your own data source
    df=df1.copy()
    mask = df["Region"] == region
    fig = px.bar(df[mask], x="country", y="%Deliverble", 
                 color="Region",template='ggplot2')
    return fig
```
